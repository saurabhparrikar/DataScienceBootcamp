## 1)Filter the data to include only weekdays (Monday to Friday) and plot a line graph showing the pedestrian counts for each day of the week.


```python
weekdays = bridge_df[(bridge_df['hour_beginning'].dt.dayofweek >= 0) & (bridge_df['hour_beginning'].dt.dayofweek) <= 4]
```

```python
weekdays
```

```python
countperday = weekdays.groupby(weekdays['hour_beginning'].dt.day_name())['Pedestrians'].sum()
```

```python
countperday
```

```python
plt.figure(figsize=(10,6))
plt.plot(countperday)
plt.title('Pedestrian Counts for Each Day of the Week')
plt.xlabel('Day of Week')
plt.ylabel('no. of Pedestrians')
plt.grid(True)
plt.show()
```

## 2) Track pedestrian counts on the Brooklyn Bridge for the year 2019 and analyze how different weather conditions influence pedestrian activity in that year. Sort the pedestrian count data by weather summary to identify any correlations( with a correlation matrix) between weather patterns and pedestrian counts for the selected year.

```python
bridge_2019 = bridge_df[(bridge_df['location'] == 'Brooklyn Bridge') & (bridge_df['hour_beginning'].dt.year == 2019)]
weather_summary_encoded = pd.get_dummies(bridge_2019['weather_summary'])



# Concatenate encoded weather summary with pedestrian counts
encoded_data = pd.concat([weather_summary_encoded, bridge_2019['Pedestrians']], axis=1)

# Calculate correlation matrix
correlation_matrix = encoded_data.corr()

# Plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Weather Summary and Pedestrian Counts')
plt.show()
```

## 3)Implement a custom function to categorize time of day into morning, afternoon, evening, and night, and create a new column in the DataFrame to store these categories. Use this new column to analyze pedestrian activity patterns throughout the day.

```python
# Custom function to categorize time of day (Intervals can be chosen according to your liking)
def categorize_time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 24:
        return 'Evening'
    else:
        return 'Night'

# Create new column to store time of day categories
bridge_df['time_of_day'] = bridge_df['hour_beginning'].dt.hour.apply(categorize_time_of_day)

# Aggregating pedestrian count throughout the day
ped_counts_per_time_of_day = bridge_df.groupby('time_of_day')['Pedestrians'].sum()

print("Pedestrian Activity Patterns Throughout the Day:")
print(ped_counts_per_time_of_day)
```
