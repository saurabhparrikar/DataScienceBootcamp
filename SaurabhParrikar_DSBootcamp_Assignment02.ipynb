```python
import pandas as pd
import numpy as np

# Create a DataFrame
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))

# Get the rows with row sum greater than 100
result = df[df.sum(axis=1) > 100]

print(result)
```

```python
# Create a 4x4 NumPy array filled with random integers between 1 and 100
arr = np.random.randint(1, 101, (4, 4))

# Reshape the array into two separate 2D arrays for rows and columns
rows = arr.reshape(2, 8)
cols = arr.T.reshape(2, 8)

# Define a lambda function to calculate the sum of each row and column
sum_row = lambda x: np.sum(x, axis=1)
sum_col = lambda x: np.sum(x, axis=1)

# Calculate the sum of each row and column using the lambda function
sum_rows_array = sum_row(rows)
sum_cols_array = sum_col(cols)

print("Rows:")
print(rows)
print("Sum of rows:")
print(sum_rows_array)
print("\nColumns:")
print(cols)
print("Sum of columns:")
print(sum_cols_array)
```
